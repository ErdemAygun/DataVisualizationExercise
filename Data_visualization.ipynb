```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = np.arange(1,6)
```

```python
y = np.arange(2,11,2)
```

```python
plt.plot(x,y,"purple")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,"blue")
plt.subplot(2,2,2)
plt.plot(y,x,"green")
plt.subplot(2,2,3)
plt.plot(y,x,"yellow")
plt.subplot(2,2,4)
plt.plot(x,x**2,"red")
plt.show()
```

```python
x = np.arange(1,6)
y = np.arange(1,11,2)
fig = plt.figure()
axes = fig.add_axes([0,0.3,0.5,0.3])
plt.show()
```

```python
x = np.arange(1,6)
y = np.arange(1,11,2)
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.3,0.4,0.3])
plt.show()
```

```python
x = np.arange(1,6)
y = np.arange(1,11,2)
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.3,0.4,0.3])

axes1.plot(y,x)
axes1.set_xlabel("Outher X")
axes1.set_ylabel("Outher Y")
axes1.set_title("Outher Graph")

axes2.plot(x,y)
axes2.set_xlabel("inner X")
axes2.set_ylabel("inner Y")
axes2.set_title("inner Graph")
plt.show()
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1)

axes[0].plot(x,y)
axes[0].set_title("First Axes")
axes[1].plot(x,x**3)
axes[1].set_title("Second Axes")

plt.tight_layout()
```

```python
fig = plt.figure(figsize = (4,3))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color = "purple")
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1, figsize = (8,6))

axes[0].plot(x,y,"red")
axes[0].set_title("First Axes")
axes[1].plot(x,x**0.5,"blue")
axes[1].set_title("Second Axes")

plt.tight_layout()
```

```python
fig = plt.figure(figsize = (6,4))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**0.5, label="x1")
axes.plot(x,x**2,label="x2")
axes.plot(x,x**3,label="x3")

axes.legend()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,linewidth = 8, linestyle = "-.",marker ="o",markersize = 13,markerfacecolor = "red",markeredgecolor = "yellow",markeredgewidth=5)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,linewidth = 8, linestyle = "-.",marker ="o",markersize = 13,markerfacecolor = "red",markeredgecolor = "yellow",markeredgewidth=5)

axes.set_xlim(0,20)
axes.set_ylim(0,35)
```
